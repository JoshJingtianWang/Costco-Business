# Google Mobility Data

```python
import pandas as pd
```

```python
google2020=pd.read_csv("2020_US_Region_Mobility_Report.csv")
google2021=pd.read_csv("2021_US_Region_Mobility_Report.csv")
google2022=pd.read_csv("2022_US_Region_Mobility_Report.csv")
```

```python
googleall=google2020.append(google2021, ignore_index=True)
```

```python
googleall=googleall.append(google2022, ignore_index=True)
```

```python
googleall.head()
```

```python
googleall = googleall.dropna(subset=['sub_region_2'])
```

```python
googleall=googleall.reset_index(drop=True)
```

```python
googleall.tail()
```

```python
googleall[.index]
```

```python
googleall.to_csv('googlemobility_US.csv')
```

```python
googleall['date'] = pd.to_datetime(googleall['date'])
```

```python
googleall['weekday'] = googleall['date'].dt.day_name()
```

```python
googleall.shape
```

```python
google_monday=googleall[googleall['weekday']=='Monday']
```

```python
google_monday.to_csv('googlemobility_US_monday.csv')
```
